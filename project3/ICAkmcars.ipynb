```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.decomposition import FastICA
from scipy.stats import kurtosis
from sklearn import metrics

%matplotlib inline
```

```python
cars_flag = True
if cars_flag:
    src = '../cars/car.data'
    import warnings
    warnings.filterwarnings(action='ignore', category=DeprecationWarning)
else:
    src = '../adult/data.txt'
data = pd.read_csv(src, index_col=0)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
# Set data set specific settings for template

mappings = {}
cols = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex','native-country', 'class']
if cars_flag:
    scoring = metrics.make_scorer(metrics.matthews_corrcoef)    
    for i in data.columns:
        data[i]=le.fit_transform(data[i])
        mappings[i] = list(le.classes_)
else:
    scoring = 'accuracy'    
    for i in cols:
        data[i]=le.fit_transform(data[i])
        mappings[i] = list(le.classes_)
```

```python
X = data.loc[:,data.columns != 'class']
y = data.iloc[:,-1]
```

```python
print(X.shape)
print(y.shape)
```

```python
comp = X.shape[1]
for x in range(comp):
    fast = FastICA(n_components = comp)
    x_trans = fast.fit_transform(X)
    #print(kurtosis(x_trans) * kurtosis(x_trans))
    print(x, ":", np.mean(kurtosis(x_trans) * kurtosis(x_trans)))
    
```

```python
X_new = FastICA(n_components=3).fit_transform(X)
```

```python
X_new
```

```python
vscore = []
sse = []
for i in range(1,8):
    km = KMeans(n_clusters=i, n_jobs=-1)
    km.fit(X_new)
    sse.append(km.score(X_new))
    y_pred = km.predict(X_new)
    vscore.append(metrics.v_measure_score(y, y_pred))
```

```python
plt.title("CARS Cluster Scores")
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.plot(range(1,8), sse)
```

```python
plt.title("CARS Cluster Scores")
plt.xlabel("Number of Clusters")
plt.ylabel("V Measure Score")
plt.plot(range(1,8), vscore)
```

```python
vscore = []
vtotals= []
sse = []
stotals = []

for n_iter in (300,600):
    for i in range(1,8):
        km = KMeans(n_clusters=i, max_iter=n_iter, n_jobs=-1)
        km.fit(X)
        sse.append(km.score(X))
        y_pred = km.predict(X)
        vscore.append(metrics.v_measure_score(y, y_pred))
    vtotals.append(vscore)
    vscore = []
    stotals.append(sse)
    sse = []
```

```python
plt.title("CARS Cluster Scores")
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.plot(range(1,8), stotals[0], label="300")
plt.plot(range(1,8), stotals[1], label="600")
plt.legend(loc='best')
```

```python
plt.title("CARS Cluster Scores")
plt.xlabel("Number of Clusters")
plt.ylabel("V Measure Score")
plt.plot(range(1,8), vtotals[0], label="300")
plt.plot(range(1,8), vtotals[1], label="600")
plt.legend(loc='best')
```
